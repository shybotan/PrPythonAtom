```python
t_csv = pd.read_csv("untitled.txt",sep='|')
t_csv.head()
```

### Step 4. Discover what is the mean age per occupation

```python
t_csv.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
occupation = set(t_csv.occupation)
data = []
for i in occupation:
    ratio = (t_csv.loc[(t_csv['occupation'] == i)&(t_csv['gender'] == 'M')])['user_id'].count()/(t_csv.loc[(t_csv['occupation'] == i)])['user_id'].count()
    data.append({'occupation': i, 'ratio': ratio})
res = pd.DataFrame(data).sort_values('ratio', ascending=False)
res 
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
r=t_csv.groupby('occupation').max()
rr=t_csv.groupby('occupation').min()
print(r['age'])
print(rr[['age']])
```

### Step 8.  For each occupation present the percentage of women and men

```python
t=t_csv[['gender','occupation']]
rr=t[(t['gender']=='F')]
r=rr.groupby('occupation').count()
print('for F')
print(r*100/t.groupby('occupation').count())
print('for M')
aa=t[(t['gender']=='M')]
a=aa.groupby('occupation').count()
print(a*100/t.groupby('occupation').count())
```
